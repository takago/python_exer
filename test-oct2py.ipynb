# oct2pyを試しに使ってみたぞ

```python
%load_ext oct2py.ipython
# この記述は冒頭に必須
```

```python
%%octave
set(0,"defaulttextfontsize",8);
set(0,"defaultaxesfontsize",3);
set(0,"defaultaxeslinewidth",0.25);
set(0,"defaultlinelinewidth",0.25);  
```

```python
%%octave -f svg
X=[1:5];
Y=X'*X
t=1:1/100:5;
plot(t,sin(t),t,cos(t))
grid on
xlabel('time')
```
